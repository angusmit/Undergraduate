## Crankâ€“Nicolson scheme for the heat equation

Consider the 1D heat equation
\begin{equation}\label{eqn:heat}
  u_t = u_{xx} \quad \textrm{ for } x \in \left( 0, 1 \right), t \in \left( 0, T \right),
\end{equation}
subject to an initial condition
\begin{equation}
  u \left( x, t = 0 \right) = \sin \left( 2 \pi x \right) + 2 \sin \left( 5 \pi x \right),
\end{equation}
and homogeneous Dirichlet boundary conditions
\begin{equation}
  u \left( x = 0, t \right) = u \left( x = 1, t \right) = 0.
\end{equation}


Recall from your lectures that the Crankâ€“Nicolson discretisation can be written as 
\begin{equation}\label{ForwardEuler}
  \frac{U_m^{n + 1} - U_m^n}{\Delta t} = \frac{1}{2} \left( \frac{U_{m - 1}^n - 2 U_m^n + U_{m + 1}^n}{(\Delta x)^2} + \frac{U_{m - 1}^{n+1} - 2 U_m^{n+1} + U_{m + 1}^{n+1}}{(\Delta x)^2} \right)  \quad \textrm{ for } m \in \left\{ 1, \ldots, M - 1 \right\}, n \in \left\{ 0, \ldots, N - 1 \right\},
\end{equation}
with

\begin{equation}\label{eqn:ic_bc}
    U_m^0  = \sin \left( 2 \pi x_m \right) + 2 \sin \left( 5 \pi x_m \right) \quad \textrm{ for } m \in \left\{ 1, \ldots, M - 1 \right\}, \\
    U_0^n  = U_M^n = 0 \quad \textrm{ for } n \in \left\{ 0, \ldots, N \right\},
\end{equation}
and where $U_m^n$ is the fully discrete approximation for
$U \left( x = x_m = m \Delta x, t = t_n = n \Delta t \right)$, with
$\Delta x = 1/M$, $\Delta t = T / N$ for some positive integers $M$, $N$.

Just like Euler methods, it is more convenient to first define $\mu = \Delta t/(\Delta x)^2$, and then at each time-step, one can implement this scheme by solving the matrix system below:

$$
A_1 \begin{bmatrix}
    U_{1}^{n+1}
    \\ U_{2}^{n+1}\\
    \vdots\\
    U_{M-1}^{n+1}
  \end{bmatrix}= A_2
  \begin{bmatrix}
    U_{1}^{n}
    \\ U_{2}^{n}\\
    \vdots\\
    U_{M-1}^{n}
  \end{bmatrix},
$$
where
$$
A_1 = \begin{bmatrix} 
    1+\mu  & -\mu/2 & 0      & \dots  & 0 \\
    -\mu/2 & 1+\mu  & -\mu/2 & \dots  & 0 \\
    \vdots & \ddots & \ddots & \ddots  &  \\
    0      & \dots  & -\mu/2 & 1+\mu  & -\mu/2 \\
    0      & \dots  &        & -\mu/2 & 1+\mu 
\end{bmatrix}
\quad
A_2 = \begin{bmatrix} 
    1-\mu  & \mu/2 & 0      & \dots  & 0 \\
    \mu/2 & 1-\mu  & \mu/2 & \dots  & 0 \\
    \vdots & \ddots & \ddots & \ddots  &  \\
    0      & \dots  & \mu/2 & 1-\mu  & \mu/2 \\
    0      & \dots  &        & \mu/2 & 1-\mu 
\end{bmatrix}.
$$



ðŸš© **Complete the script below by defining $A_1$ and $A_2$ and enforcing the boundary condition at each time step. Then run your code for $\Delta x = 0.02$, $T = 0.1$, and $\Delta t = 0.002$ and compare it with the exact solution**
\begin{equation}
  u \left( x, t \right) = e^{-4 \pi^2 t} \sin \left( 2 \pi x \right) + 2 e^{-25 \pi^2 t}\sin \left( 5 \pi x \right).
\end{equation}

Remember from the last computer lab that you may use `np.diagflat((x)*np.ones(M-1))` to put `x` in main diagonal and use `np.diagflat((y)*np.ones(M-1),-1)` to put `y` in the first diagonal below the main diagonal.

```python
'''
 This script finds a numerical approximation for the solution to
 the following 1D heat equation:
   u_t = u_xx  for  x \in (0,1),  t \in (0,T),
   u(x,t=0) = sin(2*pi*x) + 2*sin(5*pi*x),
   u(x=0,t) = u(x=1,t) = 0,
 using a Crank-Nicolson scheme.
''' 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation


## ====== setting parameters =======
# The grid spacing in the x-dimension
dx = 0.02
# The x-dimension grid locations
x = np.arange(0, 1 + dx, dx)
# The number of x-dimension grid divisions
M = np.size(x)-1
# The time step size
dt = 0.002
# The time integration length
T = 0.1
# The number of timesteps
N = np.int(np.floor(T/dt))

# Define the numerical solution 
# the first index is space and the second time
U = np.zeros((M+1,N+1))
U_exact = np.zeros((M+1,N+1))

# The initial condition
U[:,0] = np.sin(2*np.pi*x) + 2*np.sin(5*np.pi*x)
U_exact[:,0] = np.sin(2*np.pi*x) + 2*np.sin(5*np.pi*x);
# enforcing the boundary condition
U[0,0] = 0 
U[-1,0] = 0 

# Form matrix which applies the Crank-Nicolson method
mu = dt/dx**2
A_1 = np.diagflat((1+mu)*np.ones(M-1))+np.diagflat((-mu/2)*np.ones(M-2),1)+np.diagflat((-mu/2)*np.ones(M-2),-1)
A_1_inv = np.linalg.inv(A_1)
A_2 = np.diagflat((1-mu)*np.ones(M-1))+np.diagflat((mu/2)*np.ones(M-2),1)+np.diagflat((mu/2)*np.ones(M-2),-1)


for n in range(N):
    # enforcing the boundary condition
    U[0,n+1] = 0 
    U[-1,n+1] = 0 
    # A single Crank-Nicolson timestep
    U[1:M,n+1] = np.dot(A_1_inv,np.dot(A_2,U[1:M,n]))
    U_exact[:,n+1] = np.exp(-4*np.pi**2*(n+1)*dt)*np.sin(2*np.pi*x)+2*np.exp(-25*np.pi**2*(n+1)*dt)*np.sin(5*np.pi*x);
    
    

fig, ax = plt.subplots()
ax.set_xlabel(r'$x$', fontsize=18)  # Add an x-label to the axes.
ax.set_ylabel('$u$', fontsize=18)
ax.axis([0,1,-3,3])
l, = ax.plot([],[],'ok')
m, = ax.plot([],[])
def animate(i):
    l.set_data(x, U[:,i])
    m.set_data(x, U_exact[:,i])
    
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=N)

from IPython.display import HTML
HTML(ani.to_jshtml())    
```

ðŸš©**Discuss the stability of the Crankâ€“Nicolson method. What does it happen if you make $\mu$ relatively large?** 

---

**Solution**
Unlike the solution obtained from Forward Euler, the solution of this scheme does not blow-up even for large $\mu$'s. Hence, it is unconditionally stable for the heat equation. However, the approximate solution shows spurious oscillations for large values of $\mu$.

ðŸš© **Use your completed script from the previous parts to compute the two error norms below**
\begin{equation}
  \max_{n \in \left\{ 0, \ldots N \right\}} \left\| \underline{e}^n \right\|_2
    = \max_{n \in \left\{ 0, \ldots N \right\}} \sqrt{\sum_{m = 0}^M e_m^n e_m^n},
\end{equation}
and
\begin{equation}
  \max_{n \in \left\{ 0, \ldots N \right\}} \left\| \underline{e}^n \right\|_{2,\Delta x}
    = \max_{n \in \left\{ 0, \ldots N \right\}} \sqrt{\Delta x \sum_{m = 0}^M e_m^n e_m^n},
\end{equation}
where
\begin{equation}
  e_m^n = U_m^n - u \left( x = x_m = m \Delta x, t = t_n = n \Delta t \right).
\end{equation}


Then find the spatial order of accuracy of Crankâ€“Nicolson by estimating the maximum integers $p$ and $q$ such
that
\begin{equation}
  \max_{n \in \left\{ 0, \ldots N \right\}} \left\| \underline{e}^n \right\|_2 = {\cal O} \left( (\Delta x)^{p/2} \right),
\end{equation}
and
\begin{equation}
  \max_{n \in \left\{ 0, \ldots N \right\}} \left\| \underline{e}^n \right\|_{2,\Delta x} = {\cal O} \left( (\Delta x)^{q/2} \right).
\end{equation}

You can do this by plotting the norms as functions of $\Delta x$ in logscale. You can give $\Delta x$ several values such as $0.02, 0.01, 0.005$ and $0.002$, and make $\Delta t =  (\Delta x)/4 $. Use some guidelines to estimate the slope of lines in logscale. 

ðŸš© **How does the order of accuracy in space compare with Forward Euler method? Can you reason why?**

---
**Solution**
The results of the script below suggests $p = 3$ and $q = 4$, which implies this method is second order in space. We can conclude that both schemes have similar spatial order of accuracy. 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation


## ====== setting parameters =======
# The grid spacing in the x-dimension
dx_vec =np.array([0.02, 0.01, 0.005, 0.002])
max_norm2 = np.zeros_like(dx_vec);
max_norm2_dx = np.zeros_like(dx_vec);

for i, dx in enumerate(dx_vec):
    # The x-dimension grid locations
    x = np.arange(0, 1 + dx, dx)
    # The number of x-dimension grid divisions
    M = np.size(x)-1;
    # The time step size
    dt = .25*dx;
    # The time integration length
    T = 0.1;
    # The number of timesteps
    N = np.int(np.floor(T/dt))

    # Define the numerical solution 
    # the first index is space and the second time
    U = np.zeros((M+1,N+1))
    e_n = np.zeros((M+1))

    # The initial condition
    U[:,0] = np.sin(2*np.pi*x) + 2*np.sin(5*np.pi*x);
    Uexact = np.sin(2*np.pi*x) + 2*np.sin(5*np.pi*x);

    # enforcing the boundary condition
    U[0,0] = 0 
    U[-1,0] = 0 

    # Assemble matrix which applies the forward Euler method
    mu = dt/dx**2
    A_1 = np.diagflat((1+mu)*np.ones(M-1))+np.diagflat((-mu/2)*np.ones(M-2),1)+np.diagflat((-mu/2)*np.ones(M-2),-1)
    A_1_inv = np.linalg.inv(A_1)
    A_2 = np.diagflat((1-mu)*np.ones(M-1))+np.diagflat((mu/2)*np.ones(M-2),1)+np.diagflat((mu/2)*np.ones(M-2),-1)

    
    # define nornm vectors
    norm2 = np.zeros(N+1)
    norm2_dx = np.zeros(N+1)

    for n in range(N):
        # enforcing the boundary condition
        U[0,n+1] = 0 
        U[-1,n+1] = 0 
        # A single Crank-Nicolson timestep
        U[1:M,n+1] = np.dot(A_1_inv,np.dot(A_2,U[1:M,n]))
        # exact solution 
        Uexact[:] = np.exp(-4*np.pi**2*(n+1)*dt)*np.sin(2*np.pi*x)+2*np.exp(-25*np.pi**2*(n+1)*dt)*np.sin(5*np.pi*x);
        #  cacluate the error at each time step
        e_n[:] = U[:,n+1]-Uexact[:]
        norm2[n+1] = np.sqrt(np.sum(e_n**2))
        norm2_dx[n+1] = np.sqrt(dx*np.sum(e_n**2)) 

    max_norm2[i] = np.max(norm2)
    max_norm2_dx[i] = np.max(norm2_dx)

print(np.shape(norm2))
fig, ax = plt.subplots(figsize=(8, 5))    
ax.plot(dx_vec, max_norm2, label = r'$max \ || {e}^n ||_2 $')
ax.plot(dx_vec, max_norm2_dx, label = r'$max \ || {e}^n ||_{2,\Delta x} $')
# Plot the guide lines here 
ax.plot(dx_vec, 100*dx_vec**(1.5),linestyle='--', label = r'$(\Delta x)^{3/2}$')
ax.plot(dx_vec, 100*dx_vec**(2),linestyle='--', label = r'$(\Delta x)^{2}$')

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'$\Delta x$', fontsize=18)  # Add an x-label to the axes.
ax.set_ylabel('error norms', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
ax.legend(loc='lower right', fontsize = 12)
plt.show()         
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation

 


## ====== setting parameters =======
# The grid spacing in the x-dimension
dx_vec =np.array([0.02, 0.01, 0.005, 0.002])
max_norm2 = np.zeros_like(dx_vec);
max_norm2_dx = np.zeros_like(dx_vec);

 

for j, dx in enumerate(dx_vec):
    # The x-dimension grid locations
    x = np.arange(0, 1 + dx, dx)
    # The number of x-dimension grid divisions
    M = np.size(x)-1
    # The time step size
    dt = dx/4
    # The time integration length
    T = 0.1
    # The number of timesteps
    N = np.int(np.floor(T/dt))

 

    # Define the numerical solution 
    # the first index is space and the second time
    U = np.zeros((M+1,N+1))
    e_n = np.zeros((M+1))

 

    # The initial condition
    U[:,0] = np.sin(2*np.pi*x) + 2*np.sin(5*np.pi*x);
    Uexact = np.sin(2*np.pi*x) + 2*np.sin(5*np.pi*x);

 

    # enforcing the boundary condition
    U[0,0] = 0 
    U[-1,0] = 0 

 
    mu = dt/dx**2
    # Assemble matrix which applies the forward Euler method
    A_1 = np.zeros((M-1,M-1)) # Pre-allocate matrix
    for i in range(M-2):  # Fill in matrix
        A_1[i,i] = 1+mu
        A_1[i+1,i] = -mu/2
        A_1[i,i+1] = -mu/2
    A_1[M-2,M-2]=1+mu

 

    A_1_inv = np.linalg.inv(A_1)
    A_2 = np.zeros((M-1,M-1)) # Pre-allocate matrix
    for i in range(M-2):  # Fill in matrix
        A_2[i,i] = 1-mu
        A_2[i+1,i] = mu/2
        A_2[i,i+1] = mu/2
    A_2[M-2,M-2]=1-mu
 
    # define norm vectors
    norm2 = np.zeros(N+1)
    norm2_dx = np.zeros(N+1)

 

    for n in range(N):
        # enforcing the boundary condition
        U[0,n+1] = 0 
        U[-1,n+1] = 0 
        # A single Crank-Nicolson timestep
        U[1:M,n+1] = np.dot(A_1_inv,np.dot(A_2,U[1:M,n]))
        # exact solution 
        Uexact[:] = np.exp(-4*np.pi**2*(n+1)*dt)*np.sin(2*np.pi*x)+2*np.exp(-25*np.pi**2*(n+1)*dt)*np.sin(5*np.pi*x);
        #  cacluate the error at each time step
        e_n[:] = U[:,n+1]-Uexact[:]
        norm2[n+1] = np.sqrt(np.sum(e_n**2))
        norm2_dx[n+1] = np.sqrt(dx*np.sum(e_n**2)) 

 
    max_norm2[j] = np.max(norm2)
    max_norm2_dx[j] = np.max(norm2_dx)

 

fig, ax = plt.subplots(figsize=(8, 5))    
ax.plot(dx_vec, max_norm2, label = r'$max \ || {e}^n ||_2 $')
ax.plot(dx_vec, max_norm2_dx, label = r'$max \ || {e}^n ||_{2,\Delta x} $')
# Plot the guide lines here 
ax.plot(dx_vec, 100*dx_vec**(1.5),linestyle='--', label = r'$(\Delta x)^{3/2}$')
ax.plot(dx_vec, 100*dx_vec**(2),linestyle='--', label = r'$(\Delta x)^{2}$')

 

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'$\Delta x$', fontsize=18)  # Add an x-label to the axes.
ax.set_ylabel('error norms', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
ax.legend(loc='lower right', fontsize = 12)
plt.show()         
```

## Absolute stability

In this section, we want to compare the absolute stability of Forward Euler, Backward Euler and Crank-Nicolson methods. To achieve this you should first write these methods in a matrix equation

$$
\underline{U}^{n+1} = B \ \underline{U}^{n},
$$

where $\underline{U}^{n}$ is the vector of solution at time step $n$ evaluated at inner points $(m=1,2, \dots ,M-1)$ and $\underline{U}^{n+1}$ is the vector of solution at the next time step $n+1$. $B$ is a $(M-1)\times(M-1)$ matrix. 

Recall from the lecture that a method is absolutely stable if all eigenvalues $\lambda_k$ of $B$
have bounded magnitude

$$
|\lambda_k| \le 1
$$

Derive $B$ for Forward Euler, Backward Euler and Crank-Nicolson methods as a function of $\mu$. For example, using the first part of this lab for the Crank-Nicolson method $B = A_1^{-1} A_2$. Set $M=400$ and after deriving $B$ for the three methods, calculate their eigenvalues. You can do that by using `w, v = np.linalg.eig(B)` where `w` is the vector of eigenvalues and `v` the array of normalized eigenvectors (there are more efficient ways of finding eigenvalues of tridiagonal matrices but let's just use this basic numpy function). To investigate the boundedness of eigenvalues, find the max of their absolute values and plot it as a function of $\mu$ (you need a for-loop to vary $\mu$). Repeat this for Forward Euler, Backward Euler and Crank-Nicolson and discuss the results.
Change $M$ to see if your results change.

```python
import numpy as np
import matplotlib.pyplot as plt

# define a range for mu
mu_list = np.linspace(0,1,41)
# the dimension of differentiation matrices (the number of spatial grid points)
M = 400
# pre-define the list of max |\lambda_k| for different methods
max_mu_FE = []
max_mu_BE = []
max_mu_CN = []

for mu in mu_list:
    # Find the eigenvalues of Forward Euler
    B_FE = np.diagflat((1-2*mu)*np.ones(M-1))+np.diagflat((mu)*np.ones(M-2),1)+np.diagflat((mu)*np.ones(M-2),-1)
    w_FE, v_FE = np.linalg.eig(B_FE)
    # Find the eigenvalues of Backward Euler
    A = np.diagflat((1+2*mu)*np.ones(M-1))+np.diagflat((-mu)*np.ones(M-2),1)+np.diagflat((-mu)*np.ones(M-2),-1)
    B_BE = np.linalg.inv(A)
    w_BE, v_BE = np.linalg.eig(B_BE)
    # Find the eigenvalues of Crank-Nicolson
    A_1 = np.diagflat((1+mu)*np.ones(M-1))+np.diagflat((-mu/2)*np.ones(M-2),1)+np.diagflat((-mu/2)*np.ones(M-2),-1)
    A_1_inv = np.linalg.inv(A_1)
    A_2 = np.diagflat((1-mu)*np.ones(M-1))+np.diagflat((mu/2)*np.ones(M-2),1)+np.diagflat((mu/2)*np.ones(M-2),-1)
    B_CN = np.dot(A_1_inv,A_2)
    w_CN, v_CN = np.linalg.eig(B_CN)
    
    # for each method find the max of absolute values of eigenvalues
    max_mu_FE.append(max(abs(w_FE)))
    max_mu_BE.append(max(abs(w_BE)))
    max_mu_CN.append(max(abs(w_CN)))

# plot the results    
fig, ax = plt.subplots(figsize=(8, 5))        
ax.plot(mu_list, max_mu_FE, label = 'Forward Euler')
ax.plot(mu_list, max_mu_BE, linewidth=3, label = 'Backward Euler')
ax.plot(mu_list, max_mu_CN, linestyle='--', label = 'Crank-Nicolson')

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
ax.legend(loc='upper left', fontsize = 14)
plt.show()         
```
